```python
usa_rio_unique_names = usa_rio.groupby("ID").agg(lambda s: list(set(s)))
usa_rio_unique_names.head()
```

```python
usa_rio_unique_names['Sport'] = usa_rio_unique_names.Sport.apply(lambda x: x[0])
usa_rio_unique_names
```

```python
usa_rio_unique_names['Year'] = usa_rio_unique_names.Year.apply(lambda x: x[0])
usa_rio_unique_names['City'] = usa_rio_unique_names.City.apply(lambda x: x[0])
usa_rio_unique_names['Season'] = usa_rio_unique_names.Season.apply(lambda x: x[0])
usa_rio_unique_names['NOC'] = usa_rio_unique_names.NOC.apply(lambda x: x[0])
usa_rio_unique_names['Team'] = usa_rio_unique_names.Team.apply(lambda x: x[0])
usa_rio_unique_names['Weight'] = usa_rio_unique_names.Weight.apply(lambda x: x[0])
usa_rio_unique_names['Height'] = usa_rio_unique_names.Height.apply(lambda x: x[0])
usa_rio_unique_names['Age'] = usa_rio_unique_names.Age.apply(lambda x: x[0])
usa_rio_unique_names['Sex'] = usa_rio_unique_names.Sex.apply(lambda x: x[0])
usa_rio_unique_names['Name'] = usa_rio_unique_names.Name.apply(lambda x: x[0])
usa_rio_unique_names
```

```python
usa_rio_unique_names.to_csv("usa_rio_unique_names.csv")
```
