# USA Rio Data, merge medal and event

```python
usa_rio.head()
```

```python
usa_rio["Medal"] = usa_rio.Medal.fillna("Did Not Place")
```

```python
usa_rio["Sex"] = usa_rio.Sex.apply(lambda x: "Female" if x=="F" else "Male")
usa_rio.head()
```

```python
def replace_event(x):
    val = x['Event']
    gender = "Women's"
    if x["Sex"] == "Male":
        gender = "Men's"
    val = val.replace(x["Sport"], '', 1).replace(gender, '')
    return val[2:]
usa_rio["Event"] = usa_rio.apply(replace_event, axis=1)
```

```python
usa_rio["EventMedal"] = usa_rio["Event"] + ": " + usa_rio["Medal"]
```

```python
usa_rio.head()
```

```python
usa_rio_unique = usa_rio.groupby("ID").agg(lambda s: list(set(s)))
```

```python
usa_rio_unique.head()
```

```python
usa_rio_unique['Year'] = usa_rio_unique.Year.apply(lambda x: x[0])
usa_rio_unique['City'] = usa_rio_unique.City.apply(lambda x: x[0])
usa_rio_unique['Season'] = usa_rio_unique.Season.apply(lambda x: x[0])
usa_rio_unique['NOC'] = usa_rio_unique.NOC.apply(lambda x: x[0])
usa_rio_unique['Team'] = usa_rio_unique.Team.apply(lambda x: x[0])
usa_rio_unique['Weight'] = usa_rio_unique.Weight.apply(lambda x: x[0])
usa_rio_unique['Height'] = usa_rio_unique.Height.apply(lambda x: x[0])
usa_rio_unique['Age'] = usa_rio_unique.Age.apply(lambda x: x[0])
usa_rio_unique['Sex'] = usa_rio_unique.Sex.apply(lambda x: x[0])
usa_rio_unique['Name'] = usa_rio_unique.Name.apply(lambda x: x[0])
usa_rio_unique['Sport'] = usa_rio_unique.Sport.apply(lambda x: x[0])
usa_rio_unique['Games'] = usa_rio_unique.Games.apply(lambda x: x[0])

usa_rio_unique.head()
```

```python
usa_rio_unique["EventMedal"] = usa_rio_unique.EventMedal.apply(lambda x: "; ".join(x))
```

```python
usa_rio_unique.head()
```

```python
usa_rio_unique.to_csv("usa_rio_unique_names.csv")
```
